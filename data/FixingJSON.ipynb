```python
import json
```

```python
# define array to store information
hospitals_fixed = []
hospitals = []

with open('Hospital_Data.json', 'r') as openfile :
    hospitals = json.load(openfile)
```

```python
# pass information from hospital to hospital_fixed 
for idx in range(len(hospitals)) :
    idx_name = 'rs' + str(idx+1)
    hospitals_fixed.append(hospitals[idx_name])
```

```python
# write to json
with open('us_hospital_fixed.json', 'w') as outfile :
    json.dump(hospitals_fixed, outfile, indent=4)
```
